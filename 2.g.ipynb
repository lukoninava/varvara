задание5.3.

```python
df['month'] = pd.DatetimeIndex(df['date']).month
df['month'].value_counts()
def change_month(x):
    if x==4:
        return 'April'
    elif x==5:
        return 'May'
    elif x==6:
        return 'June'
    elif x==7:
        return 'Jule'
    elif x==8:
        return 'August'
    elif x==9:
        return 'September'
    elif x==10:
        return 'Oktober'
    elif x==11:
        return 'Noмember'
    elif x==12:
        return 'Dezember'
    elif x==1:
        return 'Junuar'
    elif x==2:
        return 'Februar'
    elif x==3:
        return 'Marz'
    else:
        return x

df['month_new'] = df['month'].apply(change_month)
df['month_new']
df
```
